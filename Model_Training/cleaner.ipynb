```python
import os
from pydub import AudioSegment
import tqdm

# Define the path to your train folder
train_path = 'train/'

# Define the target sample rate
target_sr = 16000



# Loop through all the subfolders in the audio directory
for class_folder_name in tqdm.tqdm(os.listdir(os.path.join(train_path, 'audio')), desc="Load all Audio and Preprocessing..."):
    class_folder_path = os.path.join(train_path, 'audio', class_folder_name)
    if os.path.isdir(class_folder_path):
        # Loop through all the audio files in the class folder
        for audio_file_name in os.listdir(class_folder_path):
            if audio_file_name.endswith('.wav'):
                # Load the audio file using pydub
                audio_path = os.path.join(class_folder_path, audio_file_name)
                audio = AudioSegment.from_file(audio_path)

                # Resample the audio to the target sample rate
                if audio.frame_rate != target_sr:
                    audio = audio.set_frame_rate(target_sr)

                # Override the original audio file
                audio.export(audio_path, format='wav')
```

```python
import os
import librosa

train_dir = "train/"

deleted_count = 0

for class_name in tqdm.tqdm(os.listdir(os.path.join(train_dir, "audio")), desc="Load all Audio and Preprocessing..."):
    class_dir = os.path.join(train_dir, "audio", class_name)
    for audio_file in os.listdir(class_dir):
        if audio_file.endswith(".wav"):
            audio_path = os.path.join(class_dir, audio_file)
            audio_length = librosa.get_duration(filename=audio_path)
            if audio_length > 1.5:
                os.remove(audio_path)
                deleted_count += 1
        else:
            os.remove(os.path.join(class_dir, audio_file))
            deleted_count += 1

print(f"Total deleted audio files: {deleted_count}")
```

```python
import zipfile
import os

# Define the path to the folder you want to zip
folder_path = '/content/train'

# Define the name and path of the zip file to be created
zip_file_path = '/content/train.zip'

# Create a new zip file
zip_file = zipfile.ZipFile(zip_file_path, 'w', zipfile.ZIP_DEFLATED)

# Walk through the files in the folder and add them to the zip file

   
for foldername, subfolders, filenames in  tqdm.tqdm(os.walk(folder_path), desc="Load all Audio and Preprocessing..."):
    for filename in filenames:
        # Construct the full path of the file
        file_path = os.path.join(foldername, filename)

        # Add the file to the zip file
        zip_file.write(file_path)

# Close the zip file
zip_file.close()

print(f"Zip file created at {zip_file_path}")
```
